```python
import re
names = []
pattern = "\"\w{10}\.csv\""
f = open("Downloads/2016.txt", "r")
for line in f:
    start = line.find("=")+2
    end = start+15
    names.append(line[start:end])
len(names)
```

```python
#!pip3 install requests
import requests
for name in names:
    url = "https://www.ncei.noaa.gov/data/local-climatological-data/access/2016/"+name
    r = requests.get(url, allow_redirects=True)
    open(name,'wb').write(r.content)
    print(name," DONE")
```
