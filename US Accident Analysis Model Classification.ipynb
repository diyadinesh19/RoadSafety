```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.nn import ReLU
from torch import sigmoid
import statistics
```

```python
losses = []
class Model(nn.Module):
    def __init__(self, hidden_units1, hidden_units2, features, features_one_hot):
        super().__init__()
        self.layer1a = nn.Linear(features, hidden_units1)
        self.layer1b = nn.Linear(features_one_hot, hidden_units1)
        self.layer2 = nn.Linear(hidden_units1*2, hidden_units2)
        #add drop out
        self.out = nn.Linear(hidden_units2, 4)
        self.loss_function = nn.MSELoss()
        self.optimizer = optim.Adam(self.parameters())
        self.dropout = nn.Dropout(0.2)
    def forward(self, x):
        r = ReLU()
        #print(x[0].shape)
        x_new = r(self.layer1a(x[0]))
        #print(self.layer1a.weight.shape,self.layer1b.weight.shape,x[0].shape,x[1].shape)
        x_new = self.dropout(x_new)
        x_one_hot = r(self.layer1b(x[1]))
        #x_one_hot = r(self.out(x[1]))
        x_one_hot = self.dropout(x_one_hot)
        x_combined = r(self.layer2(torch.cat([x_new,x_one_hot],1)))
        x_combined = self.dropout(x_combined)
        return sigmoid(self.out(x_combined))

    def train(self, data_gen, data_gen_test, epochs, batch_size):
        losses_train = []
        losses_test = []
        for epoch in range(epochs):
            avg_loss_train = 0
            avg_loss_test = 0
            i = 0
            #permutation = torch.randperm(x.size()[0])
            #for i in range(0,x.size()[0],batch_size):
            for batch_x, batch_y in data_gen:
                #indices = permutation[i:i+batch_size]
                #batch_x, batch_y = x[indices], y[indices]
                self.optimizer.zero_grad()
                prediction = self(batch_x) #x[minibatch], y[minibatch]
                loss = self.loss_function(prediction, batch_y)
                avg_loss_train+=loss.detach().numpy()
                loss.backward()
                self.optimizer.step()
                batch_x, batch_y = next(data_gen_test)
                prediction = self(batch_x) #x[minibatch], y[minibatch]
                loss = self.loss_function(prediction, batch_y)
                avg_loss_test+=loss.detach().numpy()
                
                #if i == 512*1000:
                #if i %100==0:
                #    print(i)
                if i==1000:
                    break
                i+=1
            losses_train.append(avg_loss_train/1000)
            losses_test.append(avg_loss_test/1000)
            print("avg_loss_train", avg_loss_train/1000,"avg_loss_test",avg_loss_test/1000)
            if epoch%10==0:
                with open('listfile.txt','w') as f:
                    for listitem in losses_train:
                        f.write(str(listitem))
                        f.write(" ")
                    f.write("\n")
                    f.write("\n")
                    for listitem in losses_test:
                        f.write(str(listitem))
                        f.write(" ")
                    f.write("\n")
                    f.write("\n")
                
        return losses_train, losses_test

def augment(x):
    #print("HERE",x)
    #print(x.shape)
    one, two, three, four = [], [], [], []
    for i in x.tolist():
        one.append(i[0])
        two.append(i[1])
        three.append(i[2])
        four.append(i[3])
    #print(one)
    one_v = np.var(np.array(one))/100
    #print(one_v)
    two_v = np.var(np.array(two))/100
    three_v = np.var(np.array(three))/100
    four_v = np.var(np.array(four))/100
    one_r = np.random.normal(0, one_v, size=(len(one)))#.tolist()
    #print(type(one_r),"hhh")
    two_r = np.random.normal(0, two_v, size=(len(two)))#two.shape())
    three_r = np.random.normal(0, three_v, size=(len(three)))#three.shape())
    four_r = np.random.normal(0, four_v, size=(len(four)))#four.shape())
    fin = []
    for i in range(len(one)):
        s = [one[i]+one_r[i],two[i]+two_r[i],three[i]+three_r[i],four[i]+four_r[i]]
        fin.append(s)
    #for each add a variance
    fin = torch.tensor(fin, dtype=torch.float)
    #print("NOW",fin)
    fin = fin.squeeze()
    #print(fin.shape)
    return fin
def generate_data(x, y):
    #print("hello")
    one_hot = []
    cont_x = []
    y_list = []
    for i in range(4):
        one_hot.append(x[0][y.argmax(1)==i])
        cont_x.append(x[1][y.argmax(1)==i])
        y_list.append(y[y.argmax(1)==i])
    len_list = [i.size()[0] for i in one_hot]
    def generator():
        while True:
            #perms = []
            batch_x = []
            batch_one_hot_x = []
            batch_y = []
            for index, i in enumerate(len_list):
                perm = torch.randperm(i)
                batch_x.append(augment(cont_x[index][perm[0:32]]))#cont_x[index][perm[0:32]])#augment(cont_x[index][perm[0:32]]))
                batch_one_hot_x.append(one_hot[index][perm[0:32]])
                batch_y.append(y_list[index][perm[0:32]])
                #print("hi")
            #print(y_list[1])
            #jsf
            #perm_no_accident = torch.randperm(no_accident_x.size()[0])
            batch_x = torch.cat(batch_x)
            batch_one_hot_x = torch.cat(batch_one_hot_x)
            batch_y = torch.cat(batch_y)
            #batch_y = torch.cat((accident_y[perm_accident[0:25]],no_accident_y[perm_no_accident[25:128]]))
            #batch_one_hot_x = torch.cat((one_hot_accident_x[perm_accident[0:25]],one_hot_no_accident_x[perm_no_accident[25:128]]))
            #print(batch_x.shape,batch_y.shape)
            #sgihsig
            #print(batch_x.shape, batch_one_hot_x.shape, batch_y.shape)
            yield [batch_x, batch_one_hot_x], batch_y
        
    return generator()
```
